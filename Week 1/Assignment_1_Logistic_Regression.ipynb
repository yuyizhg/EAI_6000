```python
# Step 1: Import the required modules
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
# Step 2: Preperation - import training and testing data

df_train = pd.read_csv("/Users/yuyizhang/FCR/NEU/CPS/Analytics_2018/EAI 6000_Fundamentals of Artificial Intelligence/Week 1/Titanic/train_data.csv", index_col=0)
df_test = pd.read_csv("/Users/yuyizhang/FCR/NEU/CPS/Analytics_2018/EAI 6000_Fundamentals of Artificial Intelligence/Week 1/Titanic/test_data.csv", index_col=0)
print('df_train:', df_train.shape)
print('df_test:', df_test.shape)
df_train.head()
```

```python
df_train.describe()

## From the describe result we could see taht the std value of 'PassengerId' is very high. Since this is an irrelevant variable and would cause high iterations, we are going to drop it in the next step.
```

```python
# Step 3: Preperation - assigning testing and training variables

## assign the target label 'Survived' to y_train/y_test and leave the rest columns except 'PassengerId' to x_train/x_test
x_train = df_train.drop(['PassengerId', 'Survived'], axis = 1)
y_train = df_train.Survived
x_test = df_test.drop(['PassengerId', 'Survived'], axis = 1)
y_test = df_test.Survived

## check variables' shape
print('x_train:', x_train.shape)
print('y_train:', y_train.shape)
```

```python
# Step 4: Perform Logistic Regression

## define the model
log_reg = LogisticRegression()

## fit data
log_reg.fit(x_train,y_train)

## now we are done with the model training. 
```

```python
# Step 5: Make Prediction Using the Model

## use previously prepared test data to perform the prediction
y_pred = log_reg.predict(x_test)

## find the accuracy score to evaluate performance
accuracy = metrics.accuracy_score(y_test, y_pred)
print('Accuracy Score:', accuracy)
```

```python
# Step 6: Display the Confusion Matrix

confusion_matrix(y_test, y_pred)
```
